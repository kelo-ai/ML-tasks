```python
# Iris Dataset Analysis

## Objective
The goal of this task is to explore the Iris dataset, visualize feature relationships, 
and understand patterns among different species. 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
## WE IMPORTED THE DATSET AND =LIBRARIRES FOR THE DATASET EXPLORATION
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["species"] = data.target
```

```python
## WE USE THIS STEP TO UNDERSTAND THE STRUCTURE DATATYPES AND STATISTICAL SUMMARY ABOUT THE DATA
df.head()
df.describe()
df.info()
```

```python
## VISUALIZTIONS SHOWS RELATIONSHIPS DIFFRENT DATA DISTIBUTIONS AND CORRELATIONS
# PAIRPLOT
sns.pairplot(df, hue="species")
plt.show()

# HISTOGRAM
df.hist(figsize=(8,6))
plt.show()

# HEATMAP
sns.heatmap(df.corr(), annot=True)
plt.show()
```
